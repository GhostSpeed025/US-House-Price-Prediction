# Data Prepration

```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
#Uploading CASE-SHILLER Index
df_CS = pd.read_csv("S&P.csv",names = ["Date", "S&P"], skiprows = 1)
df_CS['Date'] = pd.to_datetime(df_CS['Date'])
```

```python
df_CS
```

```python
#Resetting Index
df_CS.reset_index(inplace = True)
df_CS.drop(columns = ["index"], inplace = True)
```

```python
# Creating "Year" and "Month" columns
df_CS["Year"] = pd.DatetimeIndex(df_CS["Date"]).year
df_CS["Month"] = pd.DatetimeIndex(df_CS["Date"]).month
print(df_CS.shape)
df_CS
```

```python
# Uploading Unemployment Rate Data 
df_unemp = pd.read_csv("Unemployment Rate.csv",names = ["Date", "Unemployment Rate"], skiprows = 1)
```

```python
df_unemp
```

```python
# Uploading Per Capita GDP Data
df_pcgdp = pd.read_csv("per capita GDP.csv",names = ["Date", "Per Capita GDP"], skiprows = 1)
```

```python
df_pcgdp
```

```python
# We will be using linear interpolation. This method estimating values that fall within a range of known values 
# We will be susing it at the end 
```

```python
# Interest Rate Data
df_Fed_rate = pd.read_csv("Interest Rate.csv",names = ["Date", "Interest Rate"], skiprows = 1)
```

```python
df_Fed_rate
```

```python
# Construction material
df_cons_price_index = pd.read_csv("construction material index.csv",names = ["Date", "construction material"], skiprows = 1)
```

```python
df_cons_price_index
```

```python
# Monthly new house supply
df_house = pd.read_csv("Constructed Units.csv",names = ["Date", "New Monthly Household"], skiprows = 1)
```

```python
df_house
```

```python
# Population above 65
df_oldpop = pd.read_csv("65 above.csv", names = ["Date", "old_percent"], skiprows = 1)
```

```python
df_oldpop
```

```python
# Urban Population Percent
df_urban = pd.read_csv("Urban Population.csv",names = ["Date", "Urban Population"], skiprows = 1)
```

```python
df_urban
```

```python
# Housing Subsidies
df_subsidy = pd.read_csv("Fedral Household subsidies.csv", names = ["Date", "Subsidy"], skiprows = 1)
```

```python
df_subsidy
```

```python
# Real Median Household Income
df_income = pd.read_csv("Income.csv", names = ["Date", "Income"], skiprows = 1)
```

```python
df_income
```

```python
# Number of households
df_households = pd.read_csv("Total Household.csv", names = ["Date", "Num_Households"], skiprows = 1)
```

```python
df_households
```

```python
# Merging Per Capita GDP (Quarterly data)
df_pcgdp["Date"] = pd.to_datetime(df_pcgdp["Date"])
df_oldpop["Date"] = pd.to_datetime(df_oldpop["Date"])
df_urban["Date"] = pd.to_datetime(df_urban["Date"])
df_subsidy["Date"] = pd.to_datetime(df_subsidy["Date"])
df_households["Date"] = pd.to_datetime(df_households["Date"])


df_CS = pd.merge(df_CS,df_pcgdp, how = "left")
df_CS = pd.merge(df_CS,df_oldpop, how = "left")
df_CS = pd.merge(df_CS,df_urban, how = "left")
df_CS = pd.merge(df_CS,df_subsidy, how = "left")
df_CS = pd.merge(df_CS,df_households, how = "left")
df_CS.head()
```

```python
# Concating dataframes having monthly data to create one dataframe
df = pd.DataFrame()
df_bymonth = [df_CS, df_house, df_unemp, df_cons_price_index, df_Fed_rate,df_income]
for df1 in df_bymonth:
    df1["Date"] = pd.to_datetime(df1["Date"])
    df1 = df1.set_index("Date")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

```python
df
```

```python
df.isna().sum() # Checking the null values
```

```python
# Filling missing values in the Per_Capita_GDP column using linear interpolation
df["Per Capita GDP"] = df["Per Capita GDP"].interpolate()
```

```python
df
```

```python
df['Subsidy'].fillna(method='ffill', inplace=True)
df['old_percent'].fillna(method='ffill', inplace=True)
df['Urban Population'].fillna(method='ffill', inplace=True)
df['Num_Households'].fillna(method='ffill', inplace=True)
# Filling NaN values
```
